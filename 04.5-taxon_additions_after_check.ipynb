```python
import pandas as pd
import numpy as np
from BK_Tree import BKTree
import re
import string
from dead_processing_utils import *
from file_processing_utils import *
```

```python
determination_key = pd.read_csv("data/NoAuthorTest.csv")
determination_key = determination_key['no_author']

additions1 = pd.read_csv("output/taxon_upload/1.0-taxon_upload_FAMILIES.csv")['termDisplayName']
additions2 = pd.read_csv("output/taxon_upload/2.0-taxon_test_upload.csv")['termDisplayName']
additions3 = pd.read_csv("output/taxon_upload/3.0-taxon_test_upload_ROUND2.csv")['termDisplayName']

determination_key = pd.concat([determination_key, additions1, additions2, additions3], ignore_index=True)

determination_key = determination_key.apply(lambda x: punctStrip(x))
```

```python
checked_names_to_add = pd.read_csv('output/solved_taxon_key_additions.csv')
checked_names_to_add['checked_synonym_no_punct'] = checked_names_to_add['checked_synonym'].apply(lambda x: punctStrip(str(x)))

to_add_final = checked_names_to_add[~checked_names_to_add['checked_synonym_no_punct'].isin(determination_key)]
```
