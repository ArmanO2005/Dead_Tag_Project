Using the key that maps unique dead card collector names to correctly formatted collector names. Some of these names may already be present in the database. In this file we will determine which correctly formatted names are already present in our db and which need to be added. We will then construct a csv upload file to add those new names. 

```python
all_db_collectors = pd.read_csv('data/Collectors.csv')['0'].tolist()
dead_card_names = pd.read_csv('output\collector_name_matching_gpt_EDITED.csv')['matched_name'].dropna().unique().tolist()
```

```python
to_upload = []

for name in dead_card_names:
    if name not in all_db_collectors:
        to_upload.append(name)
```

```python
collectors_upload = strip('ucbg_templates/ucbg_4-0-0_organization-template.csv')
collectors_upload['termDisplayName'] = to_upload
collectors_upload['termStatus'] = 'accepted'
collectors_upload.to_csv('output/collectors_to_upload.csv', index=False)
```
